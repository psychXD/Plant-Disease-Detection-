```python
input_folder = './drive/MyDrive/Crop_Image_Dataset'

imageCount = 0

def process_image(input_path):
    global imageCount

    if 'BGremoved' in os.path.basename(input_path):
        return

    input_image = Image.open(input_path)
    output_image = remove(input_image)

    output_image = output_image.convert('RGB')

    filename, file_extension = os.path.splitext(os.path.basename(input_path))
    output_path = os.path.join(os.path.dirname(input_path), f'{filename}_BGremoved{file_extension}')

    output_image.save(output_path, format='JPEG')

    input_image.close()
    output_image.close()

    imageCount += 1

    os.remove(input_path)
    print(imageCount, " : Replaced file : ", input_path)

def process_images_in_directory(directory):
    for root, _, files in os.walk(directory):
        for filename in files:
            if filename.endswith(('.jpg', '.JPG', '.jpeg', '.png', '.gif', '.bmp')):
                input_path = os.path.join(root, filename)
                process_image(input_path)

if __name__ == '__main__':
    num_processes = multiprocessing.cpu_count()

    with multiprocessing.Pool(processes=num_processes) as pool:
        pool.map(process_images_in_directory, [input_folder])

    print("Background removal completed.")
```
